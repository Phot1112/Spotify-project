```python
import pandas as pd
import numpy as np
# Đọc file CSV vào một DataFrame
url="https://raw.githubusercontent.com/Phot1112/Spotify-project/e2de56e65cc13b4ca997c28544938eb25efbc333/data/raw/spotify_full_list_20102023.csv"
df = pd.read_csv(url)

#Hiển thị số dòng và cột của bộ dũ liệu
print(f"Số hàng: {df.shape[0]}")
print(f"Số cột: {df.shape[1]}")

# Hiển thị 5 dòng đầu tiên
print("5 dòng đầu tiên của DataFrame:")
df.head()
```

```python
# Hiển thị thông tin DataFrame (cột, kiểu dữ liệu, v.v.)
print("\nThông tin DataFrame:")
df.info()
```

```python
#Kiểm tra số dữ liệu bị thiếu
df.isnull().sum()
```

```python
# Xác định các cột cần điền
columns_to_fill = ['main_genre', 'genres', 'first_genre', 'second_genre', 'third_genre']

# Điền giá trị "Unknown" vào các ô trống (NaN) của những cột này
df.fillna({col: 'Unknown' for col in columns_to_fill}, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Chuyển đổi kiểu dữ liệu cột 'year'
df['year'] = df['year'].astype(int)
```

```python
#Xóa cột 'Unnamed: 0'
df.drop(columns=['Unnamed: 0'], inplace=True)
# 1. Kiểm tra số lượng dòng bị trùng lặp
duplicate_count = df.duplicated().sum()
print(f"Tìm thấy tổng cộng: {duplicate_count} dòng bị trùng lặp hoàn toàn.")
```

```python
#in ra các dòng trùng lặp
df[df.duplicated(keep=False)]
```

```python
#3. Phân tích Thống kê Mô tả
print("BƯỚC 3: PHÂN TÍCH THỐNG KÊ MÔ TẢ")
print("\n3.1. Thống kê tổng quan cho cột 'Streams'")
 # Sử dụng .to_string() để đảm bảo định dạng đẹp
print(df['Streams'].describe().to_string())
```

```python
#Phân phối của các cột phân loại
print("\n3.2. Phân phối của các cột phân loại")
        
print("\n3.2.1. Số lượng bài hát theo năm:")
print(df['year'].value_counts().sort_index().to_string())
```

```python
        
print("\n3.2.2 10 Thể loại nhạc (main_genre) hàng đầu (theo số lượng bài hát):")
print(df['main_genre'].value_counts().head(10).to_string())
```

```python
print("\n3.3. Phân tích nhóm (Trọng tâm)")

print("\n3.3.1. Top 10 Thể loại (theo Tổng Streams):")
# Tính toán và làm tròn cột 'mean' cho dễ đọc
streams_by_genre = df.groupby('main_genre')['Streams'].agg(['sum', 'mean', 'count']).sort_values('sum', ascending=False)
streams_by_genre['mean'] = streams_by_genre['mean'].round(0)
print(streams_by_genre.head(10).to_string())
```

```python
print("\n3.3.2. Top 10 Nghệ sĩ (theo Tổng Streams):")
top_artists_streams = df.groupby('Artist')['Streams'].sum().sort_values(ascending=False)
print(top_artists_streams.head(10).to_string())
```

```python

print("\n3.3.3. Top 10 Bài hát (theo Tổng Streams):")
top_songs = df.sort_values('Streams', ascending=False)[['Artist and Title', 'Artist', 'year', 'main_genre', 'Streams']]
print(top_songs.head(10).to_string())
```
